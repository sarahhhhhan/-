# BERT와 GPT
- BERT와 GPT는 트랜스포머 아키텍처를 기본으로 하는 모델이다.

- GPT(Generative Pre-trained Transformer)는 언어모델(Language Model), 이전 단어들이 주어졌을 때 다음 단어가 무엇인지 맞추는 과정에서 프리트레인(pretrain)한다. 문장 시작부터 순차적으로 계산한다는 점에서 일방향(unidirectional)으로 동작.
- BERT(Bidirectional Encoder Representations from Transformers)는 마스크 언어모델(Masked Language Model), 문장 중간에 빈칸을 만들고 해당 빈칸에 어떤 단어가 적절할지 맞추는 과정에서 프리트레인한다. 빈칸 앞뒤 문맥을 모두 살필 수 있다는 점에서 양방향(bidirectional) 성격을 가진다.
- BERT는 트랜스포머에서 인코더(encoder), GPT는 트랜스포머에서 디코더(decoder)만 취해 사용한다,

## GPT
- 트랜스포머에서 인코더를 제외하고 디코더만 사용
- 디코더 블록을 자세히 보면 인코더 쪽에서 보내오는 정보를 받는 모듈(Multi-Head Attention) 제거
image.png

- 셀프 어텐션(Masked Multi-Head Attention)과 feed forword만 사용.
> 예)입력 단어 시퀀스가 '어제 카페 갔었어 거기 사람 많더라'이고 이번이 '카페'를 맞춰야 하는 상황, 이 경우 GPT는 정답 단어 '카페'를 맞출 때 '어제'라는 단어만 참고.  
따라서 정답 단어 이후의 모든 단어('카페'~'많더라')를 볼 수 없도록 처리.  
'어제'라는 단어에 대응하는 GPT 마지막 레이어의 출력 결과에 선형 변환과 소프트맥스를 적용해 요솟값이 확률이고 학습 대상 언어의 어휘 수만큼 차원을 가진 벡터를 생성.   
'카페'의 확률을 높이고 나머지 단어의 확률은 낮아지도록 모델링.

## BERT
- 트랜스포머에서 디코더를 제외하고 인코더만 사용.
image.png

- BERT의 셀프 어텐션(Multi-Head Attention)을 도식적으로 나타낸 것.
- 입력 단어 시퀀스가 '어제 카페 갔었어 [MASK] 사람 많더라'라고 가정.
- 확인할 수 있듯 BERT는 마스크 토큰 앞뒤 문맥을 모두 참고할 수 있음.
- 앞뒤 정보를 준다고 해서 정답을 미리 알려주는 것은 아니다.
